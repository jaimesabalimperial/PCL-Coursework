```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
learning_rates = [1e-2, 5e-2, 1.5e-1, 5e-1, 1]
LR_f1_scores = [0.4684, 0.5926, 0.5855, 0.4837, 0.1736]

batch_sizes = [8, 16, 32, 64, 128]
BS_f1_scores = [0.56818, 0.57701, 0.46243 ,0.17357, 0.17357]

warmup_steps = [25, 50, 100, 500, 1000]
WS_f1_scores = [0.57325, 0.58278, 0.57391, 0.57388, 0.57516]
```

```python
fig, ax = plt.subplots(1,3, figsize=(15, 5), sharey=True)

ax[0].plot(learning_rates, LR_f1_scores)
ax[0].set_xlabel(r'Learning Rate * $10^{-4}$', fontsize=15)
ax[0].set_ylabel('f1 score', fontsize=15)

ax[1].plot(batch_sizes, BS_f1_scores)
ax[1].set_xlabel('Batch Size', fontsize=15)

ax[2].plot(warmup_steps, WS_f1_scores)
ax[2].set_xlabel('Warmup Steps', fontsize=15)

plt.subplots_adjust(wspace=0)
```
